NumPy

Primary Purpose:

NumPy (Numerical Python) is primarily used for numerical computations.
It provides support for large, multi-dimensional arrays and matrices, along with a large collection of mathematical functions to operate on these arrays.

Data Structures:

The core data structure in NumPy is the ndarray (N-dimensional array).
It supports arrays of homogeneous data types (all elements of an array must be of the same type).

Performance:

NumPy arrays are more efficient for numerical computations and are generally faster due to their use of contiguous memory and operations implemented in C.

Functionality:

NumPy is heavily used in scientific computing, linear algebra, Fourier transforms, and random number capabilities.
It includes functions for element-wise operations, broadcasting, and linear algebra operations like matrix multiplication.

NumPy ufuncs

```python
x = [1, 2, 3, 4]
y = [4, 5, 6, 7]
z = np.add(x, y)
print(x + y)
print(z)
print(type(np.add))
```

```python
# Create your own ufunc for addition:
def myadd(x, y):
  return x+y

myadd = np.frompyfunc(myadd, 2, 1)

print(myadd([1, 2, 3, 4], [5, 6, 7, 8]))
```

```python
arr1 = np.array([10, 20, 30, 40, 50, 60])
arr2 = np.array([20, 21, 22, 23, 24, 25])

newarr = np.subtract(arr1, arr2)

print(newarr)

newarr = np.multiply(arr1, arr2)

print(newarr)

newarr = np.divide(arr1, arr2)

print(newarr)

arr1 = np.array([10, 20, 30, 40, 50, 60])
arr2 = np.array([3, 5, 6, 8, 2, 33])

newarr = np.power(arr1, arr2)

print(newarr)

newarr = np.mod(arr1, arr2)

print(newarr)

newarr = np.remainder(arr1, arr2)

print(newarr)

newarr = np.divmod(arr1, arr2)

print(newarr)

arr = np.array([-1, -2, 1, 2, 3, -4])

newarr = np.absolute(arr)

print(newarr)
```

```python
# Remove the decimals, and return the float number closest to zero. Use the trunc() and fix() functions.
arr = np.trunc([-3.1666, 3.6667])

print(arr)

arr = np.fix([-3.1666, 3.6667])

print(arr)

arr = np.around(3.1666, 2)

print(arr)

arr = np.floor([-3.1666, 3.6667])

print(arr)

arr = np.ceil([-3.1666, 3.6667])

print(arr)
```

```python
# Use the log2() function to perform log at the base 2.
arr = np.arange(1, 10)
arr = np.arange(10, 0, -1)
print(arr)
print(np.log2(arr))
print(np.log10(arr))
```

```python
from math import log
import numpy as np

nplog = np.frompyfunc(log, 2, 1) # 2 is the number of input, 1 is the number of output

print(nplog((8, 16), (2, 4)))
```

```python
arr1 = np.array([1, 2, 3])
arr2 = np.array([1, 2, 3])

newarr = np.sum([arr1, arr2])
print(newarr)

newarr = np.sum([arr1, arr2], axis=1)

print(newarr)

arr = np.array([1, 2, 3])
newarr = np.cumsum(arr)

print(newarr)

arr = np.array([1, 2, 3, 4])

x = np.prod(arr)

print(x)

arr1 = np.array([1, 2, 3, 4])
arr2 = np.array([5, 6, 7, 8])

x = np.prod([arr1, arr2])

print(x)

arr = np.array([5, 6, 7, 8])

newarr = np.cumprod(arr)

print(newarr)
```

```python
# A discrete difference means subtracting two successive elements.
import numpy as np

arr = np.array([10, 15, 25, 5])

newarr = np.diff(arr)

print(newarr)

# We can perform this operation repeatedly by giving parameter n.

arr = np.array([10, 15, 25, 5])

newarr = np.diff(arr, n=2)

print(newarr)
```

```python
# The Lowest Common Multiple is the smallest number that is a common multiple of two numbers.

num1 = 4
num2 = 6

x = np.lcm(num1, num2)

print(x)

# To find the Lowest Common Multiple of all values in an array, you can use the reduce() method.
arr = np.array([3, 6, 9])

x = np.lcm.reduce(arr)

print(x)

# The GCD (Greatest Common Denominator), also known as HCF (Highest Common Factor) is the biggest number that is a common factor of both of the numbers.

num1 = 6
num2 = 9

x = np.gcd(num1, num2)
```

```python
# Convert following array with repeated elements to a set:

arr = np.array([1, 1, 1, 2, 3, 4, 5, 5, 6, 7])

x = np.unique(arr)

print(x)
```

```python
# To find the unique values of two arrays

arr1 = np.array([1, 2, 3, 4])
arr2 = np.array([3, 4, 5, 6])

newarr = np.union1d(arr1, arr2)

print(newarr)

# To find only the values that are present in both arrays
newarr = np.intersect1d(arr1, arr2, assume_unique=True)
print(newarr)

# To find only the values in the first set that is NOT present in the seconds set
newarr = np.setdiff1d(arr1, arr2, assume_unique=True)

print(newarr)

# To find only the values that are NOT present in BOTH sets
newarr = np.setxor1d(arr1, arr2, assume_unique=True)

print(newarr)
```
